#### Задание 1. Вы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде. Результаты: 1) Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550. 2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228. 3) Конверсии : conv1 = 1.05% conv2 = 1.47% .

Для оценки статистической значимости различий между конверсиями в двух группах можно использовать Z-тест. 

1. Посчитаем z-значение:

p1 = conv1 / 100 = 1.05 / 100 = 0.0105 (конверсия в первой группе)
p2 = conv2 / 100 = 1.47 / 100 = 0.0147 (конверсия во второй группе)

n1 = 15550 (число юзеров в первой группе)
n2 = 15550 (число юзеров во второй группе)

z = (p2 - p1) / sqrt(p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2)

2. Найдем критическое значение z для alpha = 5%. Для двустороннего теста, когда alpha/2 = 0.025:

z_critical = 1.96

Если |z| > z_critical, то различия являются статистически значимыми.

3. Вычислим z:

z = (0.0147 - 0.0105) / sqrt(0.0105 * (1 - 0.0105) / 15550 + 0.0147 * (1 - 0.0147) / 15550)

Если |z| > z_critical, то различия являются статистически значимыми.

4. Рассчитаем z:

z = 13.857076

Так как |z| (13.857076) > z_critical (1.96), различия в конверсиях между двумя группами являются статистически значимыми при alpha = 5%.

Рекомендация команде: Исходя из полученных результатов, эксперимент с упрощением формы заказа в магазине Утконос показывает статистически значимые различия в конверсиях между двумя группами. Сравнительно высокая конверсия во второй группе (1.47%) по сравнению с первой группой (1.05%) может указывать на то, что упрощенная форма заказа в магазине привлекательнее для пользователей и способствует большему числу покупок. Следует обсудить возможность внедрения упрощенной формы заказа в широком масштабе на основании этих результатов.

#### Задание 2. Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему ?

Для сравнения метрики конверсии в покупку между двумя группами размером 10000 элементов в каждой группе, мы можем использовать статистический критерий для долей биномиального распределения.

Наиболее подходящим статистическим критерием в данном случае будет Z-тест для разности долей, так как у нас две независимые выборки и мы хотим проверить, есть ли статистически значимая разница между долями конверсии в покупку в двух группах. Z-тест для разности долей позволяет сравнивать различия в пропорциональных данных и определить, является ли разница статистически значимой или случайной.

Для применения Z-теста необходимо выполнить следующие шаги:
1. Посчитать доли конверсии в покупку в каждой группе.
2. Вычислить разницу между долями конверсии.
3. Провести Z-тест для разности долей.

Затем, на основе полученных результатов Z-теста, мы можем сделать вывод о наличии или отсутствии статистически значимых различий в конверсии между двумя группами.
