Контрольная группа: Пользователи в этой группе продолжают использовать сервис без интеграции с ВКонтакте.

Тестовая группа: Пользователям в этой группе предлагается новая функция - интеграция с ВКонтакте.

Второстепенные метрики: Увеличение общего числа пользователей, увеличение числа новых регистраций, увеличение числа пользователей, которые возвращаются.

Если результаты будут положительными и статистически значимыми, мы можем рассмотреть возможность внедрения этой функции для всех пользователей.
